```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("company_sales_data.csv")
data.head()
```

```python
# Exercise 1: Line plit of Total profit of all months
plt.figure(figsize=(10, 6))
plt.plot(data['month_number'], data['total_profit'], marker='o', color='b', label='Total Profit')
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(data['month_number']) 
plt.grid()
plt.legend()
plt.show()
```

```python
# Exercise 2: Subplot of Bathing Soap and Facewash
plt.figure(figsize=(12, 10))

# Subplot for Bathing Soap
plt.subplot(2, 1, 1)
plt.plot(data['month_number'], data['bathingsoap'], marker='o', color='r', label='Bathing Soap Sales')
plt.title('Bathing Soap Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.xticks(data['month_number'])
plt.grid()
plt.legend()
# Subplot for Facewash
plt.subplot(2, 1, 2)
plt.plot(data['month_number'], data['facewash'], marker='o', color='g', label='Facewash Sales')
plt.title('Facewash Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Facewash Sales')
plt.xticks(data['month_number'])
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
