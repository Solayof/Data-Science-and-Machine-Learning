## Biometrics (Face, Finger, or Palm Recognition)

 Computer vision can make biometric authentication more secure and seamless:

### Face ID Payments

Customers can authorize transactions by simply looking at a camera at ATMs, POS terminals, or even mobile apps.

Example: Similar to Apple Pay’s Face ID but optimized for local needs and possibly offline validation in areas with poor connectivity.


### Anti-Spoofing Features

Computer vision can detect liveness (e.g., blinking, head movement) to prevent fraud using printed photos or deepfakes.


### Palm Vein or Fingerprint Scanning

Contactless palm scanners in banks or retail stores can reduce hygiene concerns while maintaining strong security.



Use Case:
A user walks into a shop, picks items, and walks out — cameras verify their identity and automatically deduct from their fintech wallet.


## QR-Code Payments (Enhanced with Vision)

Computer vision can upgrade QR payment experiences in several ways:

### Auto-Detection and Scanning

Customers don’t need to align their phone camera perfectly; the app can detect and read QR codes instantly from different angles or poor lighting conditions.


### Dynamic QR Recognition

Computer vision can recognize multiple QR codes in crowded spaces (e.g., markets) and prompt users to choose the right merchant.


### Fraud Detection

Vision models can detect tampered or fake QR codes — for instance, identifying altered printouts pasted over genuine merchant codes.



Use Case:
A user simply points their camera toward a store counter, and the app auto-selects the merchant QR for payment without extra taps.


## Swipe and Pay (Gesture or Object-Based Payments)

Computer vision can make gesture-driven or object-based payments a reality:

### Gesture Recognition

Customers swipe their hand, nod, or perform a gesture to authorize a payment instead of tapping buttons.


### Cardless and Contactless Payments

Cameras detect a user’s pre-registered face or hand gesture near POS terminals to approve payments, useful in high-traffic spaces.


### Item-Based Auto-Checkout

Similar to Amazon Go stores, cameras recognize the items picked by the customer and automatically calculate the bill.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# === Load the corrupted image ===
image_path = "corrupted.png"
corrupted = cv2.imread(image_path)
```

```python
# Convert from BGR to RGB for displaying
corrupted_rgb = cv2.cvtColor(corrupted, cv2.COLOR_BGR2RGB)
```

```python
# === Function to adjust brightness and contrast ===
def adjust_contrast_brightness(img, alpha=1.0, beta=0):
    """
    alpha: contrast control (1.0 = original, >1 increase contrast, <1 decrease)
    beta: brightness control (0 = original, +ve brighten, -ve darken)
    """
    adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
    return adjusted
```

```python
alpha = 0.8
beta = 20

restored = adjust_contrast_brightness(corrupted, alpha, beta)
restored_rgb = cv2.cvtColor(restored, cv2.COLOR_BGR2RGB)

# === Display results ===
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.title("Corrupted Picture")
plt.imshow(corrupted_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title(f"Restored Picture\nalpha={alpha}, beta={beta}")
plt.imshow(restored_rgb)
plt.axis('off')

plt.show()
```

```python
print(f"\nContrast number {alpha} and brightness number {beta} restored the picture to its original state.")
```
